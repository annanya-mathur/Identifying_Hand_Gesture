```python
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as pt
```

```python
img=cv2.imread('/content/hand-gestures.jpg')
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
pt.imshow(img)
```

# **MANUAL SCALING**

```python
lower=np.array([10,10,10])
upper=np.array([255,255,255])
mask=cv2.inRange(img,lower,upper)          #filter
pt.imshow(mask,'gray')
```

```python
contour,_=cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
```

```python
mask.shape
```

```python
img.shape             # 3D matrix converted to 2d matrix
```

```python
con_img=cv2.drawContours(img,contour,-1,(0,255,0),10)
pt.imshow(con_img)
```

# **MAXIMUM AREA**

```python
c=max(contour,key=cv2.contourArea)
c
```

```python
x,y,w,h=cv2.boundingRect(c)
cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),20)
```

```python
pt.imshow(img)
```

```python
cropped_img=img[y:y+h,x:x+w]
pt.imshow(cropped_img)
```
